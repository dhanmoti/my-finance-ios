```python
import numpy as np  # for numerical computation 
import pandas as pd # for data manipulation for tabular data

import matplotlib.pyplot as plt # quick data visualization
import matplotlib as mpl

import seaborn as sns # visualization with modeling information

from sklearn.preprocessing import StandardScaler # Standardization of Your Variables
from sklearn.cluster import KMeans # for K-Means clustering algorithm
from sklearn.pipeline import make_pipeline

import time # for our customized timer function

from coremltools.converters import sklearn as coreml_converter
```

```python
# Load data
data =\
(
    pd
    .read_csv("transactions.csv")
)

data.info()
```

```python
data.head(5)
```

### WRANGLE

```python
scaler = StandardScaler()

data_Z =\
(
    data
    .copy()
)
```

```python
data_Z["Amount"] =\
(
    #chaining
    #flow-programming
    scaler
    .fit_transform(data_Z
                   [["Amount"]]
                  )
)
# Drop the 'id' column
data_Z = data_Z.drop(columns=["id"])
```

```python
data_Z.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
# Label encode the 'Category' column
label_encoder = LabelEncoder()
data_Z['Category_Numeric'] = label_encoder.fit_transform(data['Category'])
data_Z.head(5)
```

```python
# Convert 'Date' to datetime
data_Z['date'] = pd.to_datetime(data_Z['date'])

# Create "Weekday" or "Weekend" column
data_Z['Day_Type'] = data_Z['date'].apply(lambda x: 'Weekend' if x.weekday() >= 5 else 'Weekday')

# Encode "Weekday" and "Weekend" numerically
data_Z['Day_Type_Numeric'] = data_Z['Day_Type'].map({'Weekday': 0, 'Weekend': 1})
data_Z.head(5)
```

```python
selected_attributes = ["Amount", "Category_Numeric", "Day_Type_Numeric"]
```

```python
K = 3

algo_searching_for_k =\
    KMeans(n_clusters = K,
           random_state = 240330)

algo_searching_for_k \
    .fit(data_Z
         [selected_attributes]
        ) # returns calculated algorithm
```

```python
algo_searching_for_k \
    .inertia_ # returns WSS, where k = 3
```

```python
WSS_scores = [] # WSS denotes Within-Clusters Sum of Squares

for K in range(2, 11):
    wss =\
        KMeans(n_clusters = K,
               random_state = 240330) \
        .fit(data_Z
             [selected_attributes]
            ) \
        .inertia_

    WSS_scores \
    .append(wss)
```

```python
WSS_scores
```

```python
plt.figure(figsize = [10, 6]
          )

plt.plot(range(2, 11),
         WSS_scores,
         "-o")

plt.title("Elbow Chart: For Deciding the Optimal N. of K")
plt.xlabel("The Number of Cluseters = K")
plt.ylabel("Within-Clusters Sum of Squares (WSS) = SSE = Inertia")
```

```python
# # Drop non-numeric columns
data_Z = data_Z.drop(columns=["Category","Description"]) 
data_Z = data_Z.drop(columns=["Day_Type"]) 
data_Z = data_Z.drop(columns=["date"]) 

data_Z.info()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import coremltools as ct


# Train the KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data_Z)

# Extract centroids
centroids = kmeans.cluster_centers_

# Add a column for cluster labels
data_Z['Cluster'] = kmeans.predict(data_Z)
data_Z.head(5)
```

```python
# print("Centroids:\n", centroids)
# print("Labeled Data:\n", data)

# Prepare a new classification dataset (centroids)
centroid_df = pd.DataFrame(centroids, columns=["Amount", "Category_Numeric", "Day_Type_Numeric", "Cluster"])
centroid_df["Cluster"] = [0, 1, 2]  # Cluster labels

centroid_df.head(5)
```

```python
# Convert centroids to CoreML for classification
# coreml_model = ct.converters.sklearn.convert(
#     centroid_df,
#     input_features=["Amount", "Category_Numeric", "Day_Type_Numeric"],
#     output_feature_names=["Cluster"]
# )
# coreml_model.info()
#---
from sklearn.neighbors import KNeighborsClassifier
# import coremltools as ct
# import pandas as pd

# Features and target
X = data_Z[["Amount", "Category_Numeric", "Day_Type_Numeric"]]
y = data_Z["Cluster"]

# Train a simple classifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)

```

```python
# Convert the trained classifier to CoreML
coreml_model = ct.converters.sklearn.convert(
    knn
)
```

```python

# Save the CoreML model
coreml_model.save("KMeansClassifier.mlmodel")
print("Model saved as KMeansClassifier.mlmodel")
```
